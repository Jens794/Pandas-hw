## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders_df = purchase_data['Item ID'].groupby(purchase_data['SN']).count()
top_spenders_df
```

```python
# Creating the initial column in the Dataframe
top_spenders_df = pd.DataFrame(data=top_spenders_df)
top_spenders_df.columns = ['Purchase Count']
top_spenders_df
```

```python
# Calculating the column values
top_spenders_df['Average Purchase Price'] = round(purchase_data['Price'].groupby(purchase_data['SN']).mean(),2)
top_spenders_df['Total Purchase Value'] = purchase_data['Price'].groupby(purchase_data['SN']).sum()
top_spenders_df
```

```python
top_spenders_df.sort_values(by=['Total Purchase Value'], ascending=False, inplace=True)
top_spenders_df
```

```python
# Formatting the currency columns into $0.00 for ease of viewing
top_spenders_df['Average Purchase Price'] = top_spenders_df['Average Purchase Price'].map('${:,.2f}'.format)
top_spenders_df['Total Purchase Value'] = top_spenders_df['Total Purchase Value'].map('${:,.2f}'.format)

top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Getting Purchase Count and Total Purchase Value
most_items_df = purchase_data.groupby(['Item ID', 'Item Name','Price'])['Price'].agg(['count','sum'])
most_items_df.columns = ['Purchase Count', 'Total Purchase Value']
most_items_df
```

```python
# Reformatting the index
most_items_df.reset_index(inplace=True)
most_items_df.set_index(['Item ID','Item Name'] ,inplace=True)
most_items_df
```

```python
# Renaming Dataframe columns
most_items_df = most_items_df[['Purchase Count', 'Price', 'Total Purchase Value']]
most_items_df
```

```python
# Sorting by most popular (Highest Purchase Count)
most_popular_items_df = most_items_df.sort_values(by='Purchase Count', ascending=False)
most_popular_items_df
```

```python
# Formatting the currency columns into $0.00 for ease of viewing
most_popular_items_df['Price'] = most_popular_items_df['Price'].map('${:,.2f}'.format)
most_popular_items_df['Total Purchase Value'] = most_popular_items_df['Total Purchase Value'].map('${:,.2f}'.format)

most_popular_items_df.head()
```
